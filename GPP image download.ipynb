```python
#Just run this with the image url csv downloaded
from pathlib import Path          
import pandas as pd
import requests
import io
from PIL import Image
```

```python
url_dict = pd.read_csv('Image_urls_GPP_new.csv', header=None, index_col=0, squeeze=True).to_dict()
url_dict
url_dict
```

```python
for key, values in url_dict.items():
    values = values.replace('[','')
    values = values.replace(']','')
    values = values.replace("'",'')
    values = values.split(',')
    n=0
    for val in values:
        n+=1
        try:
            image = requests.get(f"{val}").content
            image_file = io.BytesIO(image)
            image = Image.open(image_file).convert('RGB')

            path = Path("GPP reduced", f"{key}-{n}.jpg")
            image.save(path,'JPEG', quality = 100)
            image.close()
        except:
            continue
```
